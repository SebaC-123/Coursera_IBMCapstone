## Part 1

```python
import pandas as pd
import numpy as np
print ("Libraries imported")
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
pd.read_html(url, attrs={"class": "wikitable"})
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
df.head()
```

```python
df.replace("Not assigned", np.nan, inplace = True)
df.head(5)
```

```python
df.dropna(subset=["Borough"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df["Neighborhood"] = df["Neighborhood"].str.replace ("/", ",")
df.head()
```

```python
df.shape[0]
```

## Part 2

```python
# Load coordinate data from CSV
geo_df = pd.read_csv('http://cocl.us/Geospatial_data')
# This assumes that the postal code column in your main dataframe is labelled 'PostalCode', this renames the column from the geo csv so the merge can happen automatically and cleanly
geo_df = geo_df.rename({'Postal Code':'PostalCode'}, axis=1)
geo_df.head()
```

```python
df = pd.merge(df, geo_df, left_on='Postal code', right_on='PostalCode')
df.head()
```

```python
df = df.drop(['PostalCode'], axis=1)
df.head()
```
